## splitting a String into multiple columns


```python
df = pd.DataFrame({'name':['kashif shah','asim jan','aqib khan','arshid jaji','adil khan'],
                   'location':['peshawar,pakistan','hangu,pakistan','landi kotal,london','peshawar,India','bara road,pakistan']
                   })
```

```python
df
```

```python
df.name.str.split(' ')
```

```python
# spliting a column
df.name.str.split(' ', expand= True)
```

```python
# add those splite into new columns
df[['first_name','last_name']]= df.name.str.split(' ',expand=True)
df
```

```python
# add those splite into new columns
df[['city','country']]= df.location.str.split(',',expand=True)
df
```

```python
# refine data manipulation

df =df[['first_name','last_name','city','country']]
df
```
